### Clean the data

#### Convert data types
* name
* type
* season_availability

```python
# Convert each category into new binary
df = pd.get_dummies(df, columns=['name', 'type', 'season_availability'], drop_first=True)
```

#### Rename Columns
* freight_charge
* price_dk 
* cost_dk

```python
df.rename(columns={'freight_charge': 'freight_charge_kr', 'price_dk':'price_kr','cost_dk':'cost_kr'})
```
